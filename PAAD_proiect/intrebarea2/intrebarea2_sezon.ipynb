```python
# Summary cell: All plots from above in one script
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("/home/aureliaaa/Downloads/PAAD/PAAD/date/dataset_lunar_energie.csv")
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.month

months = [f"{m:02d}" for m in range(1, 13)]


# 3. Overlay monthly average irradiance and solar production on dual y-axes
# Chart suprapus cu ax secundar scalat pentru iradiere solar (ALLSKY_SFC_SW_DWN) i producie solar (pv_energy_kwh_day)
monthly_avg_irr = df.groupby("month")["ALLSKY_SFC_SW_DWN"].mean()
monthly_avg_pv = df.groupby("month")["pv_energy_kwh_day"].mean()
months = [f"{m:02d}" for m in range(1, 13)]
fig, ax1 = plt.subplots(figsize=(10, 5))
color1 = 'blue'
color2 = 'orange'
ax1.set_xlabel("Lun")
ax1.set_ylabel("Medie iradiere solar (ALLSKY_SFC_SW_DWN)", color=color1)
ax1.plot(months, monthly_avg_irr.values, marker='o', color=color1, label="Iradiere solar")
ax1.tick_params(axis='y', labelcolor=color1)
# Scaleaz axa secundar pentru a evidenia iradierea
ax2 = ax1.twinx()
ax2.set_ylabel("Medie producie solar (pv_energy_kwh_day)", color=color2)
ax2.plot(months, monthly_avg_pv.values, marker='o', color=color2, label="Producie solar")
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylim(min(monthly_avg_irr.values)*0.5, max(monthly_avg_pv.values)*1.2)
plt.title("Medii lunare: Iradiere solar vs Producie solar (axe scalate)")
fig.tight_layout()
plt.show()
# Valori de referin pentru plotul de iradiere solar vs producie solar
for m, irr, pv in zip(months, monthly_avg_irr.values, monthly_avg_pv.values):
    print(f"Luna {m}: Iradiere solar = {irr:.2f}, Producie solar = {pv:.2f}")


# 6. Compare monthly average temperature (T2M) with solar production
monthly_avg_temp = df.groupby("month")["T2M"].mean()
fig, ax1 = plt.subplots(figsize=(10, 5))
color1 = 'blue'
color2 = 'orange'
ax1.set_xlabel("Month")
ax1.set_ylabel("Avg Temperature (T2M, %)", color=color1)
ax1.plot(months, monthly_avg_temp.values, marker='o', color=color1, label="Temperature")
ax1.tick_params(axis='y', labelcolor=color1)
ax2 = ax1.twinx()
ax2.set_ylabel("Avg Solar Production (pv_energy_kwh_day)", color=color2)
ax2.plot(months, monthly_avg_pv.values, marker='o', color=color2, label="Solar Production")
ax2.tick_params(axis='y', labelcolor=color2)
plt.title("Monthly Average: Temperature (T2M) vs Solar Production (pv_energy_kwh_day)")
fig.tight_layout()
plt.show()
# Valori de referin pentru plotul de temperatur vs producie solar
for m, temp, pv in zip(months, monthly_avg_temp.values, monthly_avg_pv.values):
    print(f"Month {m}: Temperature = {temp:.2f}, Solar Production = {pv:.2f}")



# 8. Monthly average air density and total wind energy
R_specific = 287.05
monthly_avg_ps = df.groupby("month")["PS"].mean()
monthly_avg_temp = df.groupby("month")["T2M"].mean()
monthly_air_density = (monthly_avg_ps * 1000) / (R_specific * (monthly_avg_temp + 273.15))
monthly_total_wind = df.groupby("month")["wind_energy_kwh_day"].sum()
fig, ax1 = plt.subplots(figsize=(10, 5))
color1 = 'teal'
color2 = 'navy'
ax1.set_xlabel("Month")
ax1.set_ylabel("Air Density (kg/m)", color=color1)
ax1.plot(months, monthly_air_density.values, marker='o', color=color1, label='Air Density')
ax1.tick_params(axis='y', labelcolor=color1)
ax2 = ax1.twinx()
ax2.set_ylabel("Total Wind Energy (kWh)", color=color2)
ax2.plot(months, monthly_total_wind.values, marker='s', color=color2, label='Total Wind Energy')
ax2.tick_params(axis='y', labelcolor=color2)
plt.title("Monthly Average Air Density vs Total Eolian Energy Production")
fig.tight_layout()
plt.show()
# Valori de referin pentru plotul de densitate aer vs energie eolian
for m, val, wind in zip(months, monthly_air_density.values, monthly_total_wind.values):
    print(f"Month {m}: Air Density = {val:.3f} kg/m, Total Wind Energy = {wind:.2f} kWh")

# 10. Compare monthly average humidity (RH2M) with wind energy production
monthly_avg_humidity = df.groupby("month")["RH2M"].mean()
monthly_avg_wind = df.groupby("month")["wind_energy_kwh_day"].mean()
fig, ax1 = plt.subplots(figsize=(10, 5))
color1 = 'blue'
color2 = 'navy'
ax1.set_xlabel("Month")
ax1.set_ylabel("Avg Humidity (RH2M, %)", color=color1)
ax1.plot(months, monthly_avg_humidity.values, marker='o', color=color1, label="Humidity")
ax1.tick_params(axis='y', labelcolor=color1)
ax2 = ax1.twinx()
ax2.set_ylabel("Avg Wind Energy (wind_energy_kwh_day)", color=color2)
ax2.plot(months, monthly_avg_wind.values, marker='o', color=color2, label="Wind Energy")
ax2.tick_params(axis='y', labelcolor=color2)
plt.title("Monthly Average: Humidity (RH2M) vs Wind Energy Production (wind_energy_kwh_day)")
fig.tight_layout()
plt.show()
# Valori de referin pentru plotul de umiditate vs energie eolian
for m, hum, wind in zip(months, monthly_avg_humidity.values, monthly_avg_wind.values):
    print(f"Month {m}: Humidity = {hum:.2f}, Wind Energy = {wind:.2f}")

# 12. Clustered bar chart: Monthly mean for solar and eolian energy metrics
solar_metrics = ["ALLSKY_SFC_SW_DWN", "CLOUD_AMT", "PRECTOTCORR_SUM", "T2M", "RH2M", "pv_energy_kwh_day"]
eolian_metrics = ["WS50M", "PS", "T2M", "RH2M", "wind_energy_kwh_day"]
monthly_solar = df.groupby("month")[solar_metrics].mean()
monthly_eolian = df.groupby("month")[eolian_metrics].mean()
labels = [f"Month {m}" for m in months]
x = np.arange(len(months))
width = 0.35
fig, ax = plt.subplots(figsize=(18, 7))
for i, metric in enumerate(solar_metrics):
    ax.bar(x - width/2 + i*(width/len(solar_metrics)), monthly_solar[metric], width/len(solar_metrics), label=metric)
ax.set_ylabel("Mean Value (Solar)")
ax.set_title("Clustered Bar Chart: Monthly Mean Solar Energy Metrics [Log Scale]")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.set_yscale('log')
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
# Valori de referin pentru fiecare metric solar
for metric in solar_metrics:
    print(f"--- {metric} ---")
    for m, val in zip(months, monthly_solar[metric]):
        print(f"Month {m}: {metric} = {val:.2f}")

fig, ax = plt.subplots(figsize=(18, 7))
for i, metric in enumerate(eolian_metrics):
    ax.bar(x - width/2 + i*(width/len(eolian_metrics)), monthly_eolian[metric], width/len(eolian_metrics), label=metric)
ax.set_ylabel("Mean Value (Eolian)")
ax.set_title("Clustered Bar Chart: Monthly Mean Eolian Energy Metrics [Log Scale]")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.set_yscale('log')
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
# Valori de referin pentru fiecare metric eolian
for metric in eolian_metrics:
    print(f"--- {metric} ---")
    for m, val in zip(months, monthly_eolian[metric]):
        print(f"Month {m}: {metric} = {val:.2f}")

fig, ax = plt.subplots(figsize=(18, 7))
for metric in solar_metrics:
    ax.plot(months, monthly_solar[metric], marker='o', label=metric)
ax.set_ylabel("Mean Value (Solar)")
ax.set_title("Time Series: Monthly Mean Solar Energy Metrics")
ax.set_xticks(months)
ax.set_xticklabels(labels, rotation=45)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
# Valori de referin pentru time series solar
for metric in solar_metrics:
    print(f"--- {metric} ---")
    for m, val in zip(months, monthly_solar[metric]):
        print(f"Month {m}: {metric} = {val:.2f}")

fig, ax = plt.subplots(figsize=(18, 7))
for metric in eolian_metrics:
    ax.plot(months, monthly_eolian[metric], marker='o', label=metric)
    ax.set_ylim(-20, 150)
ax.set_ylabel("Mean Value (Eolian)")
ax.set_title("Time Series: Monthly Mean Eolian Energy Metrics")
ax.set_xticks(months)
ax.set_xticklabels(labels, rotation=45)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
# Valori de referin pentru time series eolian
for metric in eolian_metrics:
    print(f"--- {metric} ---")
    for m, val in zip(months, monthly_eolian[metric]):
        print(f"Month {m}: {metric} = {val:.2f}")

# 14. Suprapunere: Producerea total de energie solar i eolian (sum lunar)
monthly_total_pv = df.groupby("month")["pv_energy_kwh_day"].sum()
monthly_total_wind = df.groupby("month")["wind_energy_kwh_day"].sum()
fig, ax1 = plt.subplots(figsize=(10, 5))
color1 = 'orange'
color2 = 'navy'
ax1.set_xlabel("Lun")
ax1.set_ylabel("Total energie solar (kWh)", color=color1)
ax1.plot(months, monthly_total_pv.values, marker='o', color=color1, label='Solar')
ax1.tick_params(axis='y', labelcolor=color1)
ax2 = ax1.twinx()
ax2.set_ylabel("Total energie eolian (kWh)", color=color2)
ax2.plot(months, monthly_total_wind.values, marker='s', color=color2, label='Eolian')
ax2.tick_params(axis='y', labelcolor=color2)
plt.title("Producerea total lunar: energie solar vs eolian")
fig.tight_layout()
plt.show()
# Valori de referin pentru total lunar solar vs eolian
for m, pv, wind in zip(months, monthly_total_pv.values, monthly_total_wind.values):
    print(f"Luna {m}: Solar = {pv:.2f} kWh, Eolian = {wind:.2f} kWh")
```
